```python
import pandas as pd
```

```python
df = pd.read_csv("WorldEducation.csv",index_col = 0)
```

```python
df.info()
```

** There's plenty of countries + indicators to choose from, so let's see if we can focus on the categories that have the most data. **

```python
# get all possible countries
countries = df['Country Code'].unique()
# get all possible Indicators
indicators = df['Indicator Code'].unique()
```

```python
print(len(countries),len(indicators))
```

```python
# extract the numeric column names
years = df.columns[4:]
years
```

```python
# count number of non nulls 
df['Non Null'] = df[years].count(axis = 1)
```

```python
# take a look at each # non nulls for country/indicator pairs
df_temp = df[[col for col in df.columns if col not in years]]
df_temp
```

```python
# sort this dataframe based on most data
df_temp.sort_values('Non Null',ascending=False,inplace=True)
```

```python
# extract sorted indices and sort og dataframe based on this
sorted_id = df_temp.index
df = df.iloc[sorted_id]
```

```python
df.head()
```

** From this quick look, it looks like Kyrgyz Republic (KGZ) might have some interesting info **

```python
df_kgz = df.loc[df['Country Code']=='KGZ']
```

```python
# extract only indicators > 30 non nulls
df_kgz = df_kgz.loc[df_kgz['Non Null']>30]
```

```python
# take a look at what's left
df_kgz[['Indicator Name','Non Null']]
```

To do a simplified example, let's take a look at how **population** changes over the years in addition to male/female %

```python
desired = ['Population, total','Population, male','Population, female']
df_mini = df_kgz[df['Indicator Name'].isin(desired)]
```

```python
df_mini
```

```python
# drop the years that don't have data
df_mini.dropna(axis = 1,how = 'any',inplace=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = 12, 7
```

```python
years = df_mini.columns[4:]
```

```python
# setup x and y axis
x = np.linspace(1,len(years),len(years))
yt = df_mini.iloc[0][years]
ym = df_mini.iloc[1][years]
yf = df_mini.iloc[2][years]
# set up tickmarks
locs = np.arange(1,len(years), step=2)
plt.xticks(locs,[years[i] for i in range(len(years)) if i%2 == 0],rotation = 35)
# labels
plt.xlabel('Year')
plt.ylabel('Population')
plt.plot(x,yt)
plt.title('Population of KGZ 1977-2015')
p1 = plt.bar(x, ym, color='#d62728')
p2 = plt.bar(x, yf, bottom=ym)
plt.legend((p2[0],p1[0]),['Female Population','Male Population'])
```

Taking a quick glance at the trend, it looks like both males and females increase at about the same rate. There was some flat growth from 1992-1995, so that might warrant further analysis on why that may have happened!
